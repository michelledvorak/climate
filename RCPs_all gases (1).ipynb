```python
%matplotlib inline
import fair
import numpy as np
from matplotlib import pyplot as plt
from fair.forward3 import fair_scm
```

```python
from fair.RCPs import rcp3pd, rcp45, rcp6, rcp85
```

```python

C26, F26, T26 = fair.forward.fair_scm(emissions=rcp3pd.Emissions.emissions)
C45, F45, T45 = fair.forward.fair_scm(emissions=rcp45.Emissions.emissions)
C60, F60, T60 = fair.forward.fair_scm(emissions=rcp6.Emissions.emissions)
C85, F85, T85 = fair.forward.fair_scm(emissions=rcp85.Emissions.emissions)

C26_2, F26_2, T26_2, _ = fair.forward3.fair_scm(emissions=rcp3pd.Emissions.emissions, eps=1.28, gam=0.963/1.28)
C45_2, F45_2, T45_2, _ = fair.forward3.fair_scm(emissions=rcp45.Emissions.emissions, eps=1.28, gam=0.963/1.28)
C60_2, F60_2, T60_2, _ = fair.forward3.fair_scm(emissions=rcp6.Emissions.emissions, eps=1.28, gam=0.963/1.28)
C85_2, F85_2, T85_2, _ = fair.forward3.fair_scm(emissions=rcp85.Emissions.emissions, eps=1.28, gam=0.963/1.28)
```

```python
plt.style.use('seaborn-darkgrid')
plt.rcParams['figure.figsize'] = (16,12)

fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

ax1.plot(rcp3pd.Emissions.year, rcp3pd.Emissions.co2_fossil, color='green', label='RCP3PD')
ax2.plot(rcp3pd.Emissions.year, C26[:, 0], color='green')
ax2.plot(rcp3pd.Emissions.year, C26_2[:, 0], color='chartreuse')
ax3.plot(rcp3pd.Emissions.year, np.sum(F26, axis=1), color='green')
ax3.plot(rcp3pd.Emissions.year, np.sum(F26_2, axis=1), color='chartreuse')
ax4.plot(rcp3pd.Emissions.year, T26, color='green', label='FaIR')
ax4.plot(rcp3pd.Emissions.year, T26_2, color='chartreuse', label='GeoV1')

ax1.plot(rcp45.Emissions.year, rcp45.Emissions.co2_fossil, color='blue', label='RCP4.5')
ax2.plot(rcp45.Emissions.year, C45[:, 0], color='blue')
ax2.plot(rcp45.Emissions.year, C45_2[:, 0], color='lightblue')
ax3.plot(rcp45.Emissions.year, np.sum(F45, axis=1), color='blue')
ax3.plot(rcp45.Emissions.year, np.sum(F45_2, axis=1), color='lightblue')
ax4.plot(rcp45.Emissions.year, T45, color='blue', label='FaIR')
ax4.plot(rcp45.Emissions.year, T45_2, color='lightblue', label='GeoV1')

ax1.plot(rcp6.Emissions.year, rcp6.Emissions.co2_fossil, color='red', label='RCP6')
ax2.plot(rcp6.Emissions.year, C60[:, 0], color='red')
ax2.plot(rcp6.Emissions.year, C60_2[:, 0], color='pink')
ax3.plot(rcp6.Emissions.year, np.sum(F60, axis=1), color='red')
ax3.plot(rcp6.Emissions.year, np.sum(F60_2, axis=1), color='pink')
ax4.plot(rcp6.Emissions.year, T60, color='red', label='FaIR')
ax4.plot(rcp6.Emissions.year, T60_2, color='pink', label='GeoV1')

ax1.plot(rcp85.Emissions.year, rcp85.Emissions.co2_fossil, color='black', label='RCP8.5')
ax2.plot(rcp85.Emissions.year, C85[:, 0], color='black')
ax2.plot(rcp85.Emissions.year, C85_2[:, 0], color='gray')
ax3.plot(rcp85.Emissions.year, np.sum(F85, axis=1), color='black')
ax3.plot(rcp85.Emissions.year, np.sum(F85_2, axis=1), color='gray')
ax4.plot(rcp85.Emissions.year, T85, color='black', label='FaIR')
ax4.plot(rcp85.Emissions.year, T85_2, color='gray', label='GeoV1')

ax1.set_ylabel('Fossil CO$_2$ Emissions (GtC)')
ax1.legend()
ax4.legend()
ax2.set_ylabel('CO$_2$ concentrations (ppm)')
ax3.set_ylabel('Total radiative forcing (W.m$^{-2}$)')
ax4.set_ylabel('Temperature anomaly (K)');
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)

ax1.plot(rcp3pd.Emissions.year, C26_2[:,1], color='green', label='RCP3PD')
ax1.plot(rcp45.Emissions.year, C45_2[:,1], color='blue', label='RCP4.5')
ax1.plot(rcp6.Emissions.year, C60_2[:,1], color='red', label='RCP6')
ax1.plot(rcp85.Emissions.year, C85_2[:,1], color='black', label='RCP8.5')
ax1.set_title("Methane concentrations, ppb")

ax2.plot(rcp3pd.Emissions.year, C26_2[:,2], color='green', label='RCP3PD')
ax2.plot(rcp45.Emissions.year, C45_2[:,2], color='blue', label='RCP4.5')
ax2.plot(rcp6.Emissions.year, C60_2[:,2], color='red', label='RCP6')
ax2.plot(rcp85.Emissions.year, C85_2[:,2], color='black', label='RCP8.5')
ax2.set_title("Nitrous oxide concentrations, ppb")

```
